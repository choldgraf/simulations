```python
# vars
n_sig = 50
srate = 500
ntime = srate * 2
# Create random data
data = np.random.randn(n_sig, ntime)
time = np.arange(ntime)
var = 10

# Define a "bump" in each trial
activation = 1000*st.norm.pdf(time, ntime/2, var)
signal = activation + data

# Take the TFR of each trial with MNE
info = mne.create_info(['sig'], srate)
events = np.random.randint(0, 100, (n_sig, 3))
epochs = mne.EpochsArray(signal[:, None, :], info, events)

# See what the frequency content looks like
psd, fs = mne.time_frequency.multitaper_psd(epochs.get_data().squeeze(), sfreq=srate, fmax=25)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(epochs.get_data().squeeze().mean(0))
ax2.plot(fs, psd[4])
```
